```python
import matplotlib.pyplot as plt

# Data points
price = [1, 2]
quantity = [6, 3]

# Create a plot
plt.plot(quantity, price, marker='o', linestyle='-', color='blue')

# Add labels
plt.xlabel('Quantity')
plt.ylabel('Price')

# Set the title
plt.title('Demand Curve')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Define the demand curve equation
def demand_curve(quantity):
    return -3 * quantity + 9

# Generate a range of quantities
quantities = range(0, 4)

# Calculate the corresponding prices
prices = [demand_curve(quantity) for quantity in quantities]

# Create a plot
plt.plot(quantities, prices, marker='o', linestyle='-', color='blue')

# Add labels
plt.xlabel('Quantity')
plt.ylabel('Price')

# Set the title
plt.title('Demand Curve')

# Show the plot
plt.show()
```
